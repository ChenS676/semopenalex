```python
from matplotlib import pyplot as plt
from matplotlib.figure import Figure
import seaborn as sns
#sns.set_theme()
import matplotlib
matplotlib.rcParams['font.family'] = 'serif'

EXPORT_PATH = r'..\pycharm_exports'

font = {'family': 'serif',
        'color':  'black',
        'weight': 200,
        'size': 16,
        }
```

```python
# MRR single
x = [1,2,3]
gat = [0.63156,0.621569,0.657205]
gnn = [0.659639,0.679222,0.688465]
transe = [0.372694,0.378573,0.371866]
complex = [0.650792,0.691778,0.692951]
distmult= [0.659887,0.694717,0.695431]


fig = plt.figure(figsize = (4, 3))
ax = fig.add_axes([1, 1, 1, 1])
ax.step
ax1 = ax.plot(x, distmult, marker = 's', ms = 8, linewidth=2, linestyle ='dashed' ,label='DistMult')
ax2 = ax.plot(x, complex, marker = '^', ms = 8, linewidth=2, linestyle ='dotted', label='ComplEx')
ax3 = ax.plot(x, gnn, marker = 'd', ms = 8, linewidth=2, linestyle ='dotted', label='GNN')
ax4 = ax.plot(x, gat, marker = 'v', ms = 8, linewidth=2, linestyle ='dashed', label='GAT')
ax5 = ax.plot(x, transe, marker = 'p', ms = 8, linewidth=2, linestyle ='dashed', label='TransE')


plt.xlim(0.9, 3.1)
plt.xticks([1,2,3], fontsize=font['size'])
plt.yticks([0.3, 0.4, 0.5, 0.6, 0.7, 0.8],fontsize=font['size'])
plt.ylim(0.3, 0.85)
plt.xlabel('Epoch', fontdict=font)
plt.ylabel('MRR', fontdict=font)
plt.grid(color=(0.85,0.85,0.85))
#ax.legend(loc="right", bbox_to_anchor=(1.28,0.5),prop={'size': 14},facecolor='white',edgecolor='white')

plt.savefig(EXPORT_PATH+"\MRR_dev_epochs_2.pdf", format="pdf", bbox_inches="tight")
plt.show()
```
